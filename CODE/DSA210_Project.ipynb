```python
# Correlation matrix
correlation_matrix = cleaned_data[['Tempo', 'Loudness', 'Productivity']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix: Tempo, Loudness, and Productivity")
plt.show()
```

```python
# Convert Date to datetime format
cleaned_data['Date'] = pd.to_datetime(cleaned_data['Date'], dayfirst=True)

# Plot productivity trends over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Productivity', data=cleaned_data, ci=None, marker='o')
plt.title("Productivity Trends Over Time")
plt.xticks(rotation=45)
plt.show()
```

```python
# Extract the top genres contributing to productivity
genre_productivity = cleaned_data.groupby('Genre')['Productivity'].mean().sort_values(ascending=False).head(10)

# Bar plot for top genres
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_productivity.values, y=genre_productivity.index)
plt.title("Top Genres by Average Productivity")
plt.xlabel("Average Productivity")
plt.ylabel("Genre")
plt.show()
```

```python
#Outlier Detection and Removal
plt.figure(figsize=(10, 5))
sns.boxplot(data=cleaned_data[['Tempo', 'Loudness', 'Productivity']])
plt.title("Boxplots for Tempo, Loudness, and Productivity")
plt.show()

# Remove outliers
cleaned_data = cleaned_data[cleaned_data['Tempo'].between(60, 200)]
cleaned_data = cleaned_data[cleaned_data['Loudness'].between(-5, 5)]

# Remove rows with Productivity = 0
cleaned_data = cleaned_data[cleaned_data['Productivity'] > 0]
```

```python
from sklearn.preprocessing import StandardScaler
# Interaction feature
cleaned_data['Tempo_Loudness_Interaction'] = cleaned_data['Tempo'] * cleaned_data['Loudness']

# Preprocessing: Target encoding for Genre
genre_target_mean = cleaned_data.groupby('Genre')['Productivity'].mean()
cleaned_data['Genre_Target_Encoding'] = cleaned_data['Genre'].map(genre_target_mean)

# Feature matrix and target variable
X = cleaned_data[['Tempo', 'Loudness', 'Genre_Target_Encoding', 'Tempo_Loudness_Interaction']]
y = cleaned_data['Productivity']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
#Baseline Model
baseline_pred = [y.mean()] * len(y)
baseline_rmse = mean_squared_error(y, baseline_pred, squared=False)
print("Baseline RMSE:", baseline_rmse)
```

```python
from sklearn.feature_selection import mutual_info_regression
# Feature Importance
feature_importance = mutual_info_regression(X, y)
importance_df = pd.DataFrame({
    'Feature': ['Tempo', 'Loudness', 'Genre_Target_Encoding', 'Tempo_Loudness_Interaction'],
    'Importance': feature_importance
}).sort_values(by='Importance', ascending=False)
print("Feature Importance:\n", importance_df)
```

```python
# Visualize Feature Importance
sns.barplot(x='Importance', y='Feature', data=importance_df)
plt.title("Feature Importance")
plt.show()
```

```python
# Scatterplots for Key Features
sns.scatterplot(x='Tempo', y='Productivity', data=cleaned_data)
plt.title("Tempo vs Productivity")
plt.show()

sns.scatterplot(x='Loudness', y='Productivity', data=cleaned_data)
plt.title("Loudness vs Productivity")
plt.show()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error

#Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

#Hyperparameter Tuning for Random Forest
param_grid = {'n_estimators': [50, 100], 'max_depth': [3, 5]}
grid_search = GridSearchCV(RandomForestRegressor(), param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best Parameters:", grid_search.best_params_)

# Train tuned Random Forest
tuned_rf = RandomForestRegressor(max_depth=3, n_estimators=50, random_state=42)
tuned_rf.fit(X_train, y_train)
y_pred_rf = tuned_rf.predict(X_test)

# Evaluate Random Forest
print("Tuned Random Forest R²:", r2_score(y_test, y_pred_rf))
print("Tuned Random Forest RMSE:", mean_squared_error(y_test, y_pred_rf, squared=False))
```

```python
from sklearn.cluster import KMeans
#Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
cleaned_data['Cluster'] = kmeans.fit_predict(X_scaled)

# Cluster Summary
cluster_summary = cleaned_data.groupby('Cluster').agg(
    Avg_Productivity=('Productivity', 'mean'),
    Avg_Tempo=('Tempo', 'mean'),
    Avg_Loudness=('Loudness', 'mean'),
    Avg_Genre_Encoding=('Genre_Target_Encoding', 'mean')
)
print(cluster_summary)
```

```python
# Refine Clustering Insights
for cluster in cleaned_data['Cluster'].unique():
    print(f"Cluster {cluster} Characteristics:")
    print(cleaned_data[cleaned_data['Cluster'] == cluster].describe())

sns.boxplot(x='Cluster', y='Productivity', data=cleaned_data)
plt.title("Productivity Distribution by Cluster")
plt.show()
```

```python
# Correlation Analysis within Clusters
for cluster in cleaned_data['Cluster'].unique():
    cluster_data = cleaned_data[cleaned_data['Cluster'] == cluster]
    print(f"Cluster {cluster} Correlations:")
    print(cluster_data[['Tempo', 'Loudness', 'Productivity']].corr())
```

```python
#Visualize Clusters
sns.scatterplot(x='Tempo', y='Loudness', hue='Cluster', data=cleaned_data)
plt.title("Clustering by Tempo and Loudness")
plt.show()

#Pairplot for cluster characteristics
sns.pairplot(cleaned_data, hue='Cluster', vars=['Tempo', 'Loudness', 'Productivity'])
plt.show()
```

```python
# Generate and Interpret Recommendations
print("Cluster-Based Recommendations:")
for cluster in cluster_summary.index:
    avg_productivity = cluster_summary.loc[cluster, 'Avg_Productivity']
    avg_tempo = cluster_summary.loc[cluster, 'Avg_Tempo']
    avg_loudness = cluster_summary.loc[cluster, 'Avg_Loudness']
    avg_genre_encoding = cluster_summary.loc[cluster, 'Avg_Genre_Encoding']
    
    print(f"\nCluster {cluster}:")
    print(f" - Average Productivity: {avg_productivity:.2f}")
    print(f" - Recommended Tempo: {avg_tempo:.2f} BPM")
    print(f" - Recommended Loudness: {avg_loudness:.2f} (normalized scale)")
    print(f" - Genre Preference Score: {avg_genre_encoding:.2f}")
    print("Recommendation: Focus on tracks with these attributes to maximize productivity.")
```
